```python
from pathlib import Path
import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read the csv file into a pandas DataFrame
df_tx = pd.read_csv(Path("../../../data/processed_data/tx_cleaned_ev_stations_data.csv"))
```

```python
# Group the data by county and calculate the total number of EV stations for each county
county_data = df_tx.groupby('county')['ev_station_count'].sum()

# Create a bar chart to visualize the number of EV stations by county
county_data.plot(kind='bar', figsize=(10, 6))
plt.xlabel('County')
plt.ylabel('Number of EV Stations')
plt.title('Number of EV Stations by County')
plt.show()

# Create a pie chart to visualize the distribution of EV stations by county
county_data.plot(kind='pie', figsize=(10, 6), autopct='%1.1f%%')
plt.title('Distribution of EV Stations by County')
plt.show()
```
