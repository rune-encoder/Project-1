```python
import pandas as pd
from pathlib import Path
import datetime as dt
import numpy as np
```

```python
#Import Raw Median Income by County data
fl_raw_18_df = pd.read_csv(Path("../../../../data/raw_data/median_income/ACSST5Y2018.S1903-2024-09-04T005331.csv"))
fl_raw_19_df = pd.read_csv(Path("../../../../data/raw_data/median_income/ACSST5Y2019.S1903-2024-09-04T005419.csv"))
fl_raw_20_df = pd.read_csv(Path("../../../../data/raw_data/median_income/ACSST5Y2020.S1903-2024-09-04T005442.csv"))
fl_raw_21_df = pd.read_csv(Path("../../../../data/raw_data/median_income/ACSST5Y2021.S1903-2024-09-04T005457.csv"))
```

**2018 Data**

```python
#2018 Data
print("Shape:",fl_raw_18_df.shape)
display(fl_raw_18_df.columns)
display(fl_raw_18_df.head(10))
```

```python
#Rename 2018 Columns
fl_18_df = fl_raw_18_df.rename(columns={"Label (Grouping)":"County","HOUSEHOLD INCOME BY RACE AND HISPANIC OR LATINO ORIGIN OF HOUSEHOLDER!!Households":"2018 Median Income"})
fl_18_df.head(20)
```

```python
#Moving County names
fl_18_df['county'] = fl_18_df['County'].where(fl_18_df['County'].str.contains('County'))

fl_18_df['county'] = fl_18_df['county'].ffill()

fl_18_df = fl_18_df.loc[~fl_18_df['County'].str.contains('County')]

fl_18_df.head(15)
```

```python
#Moving values
filter_number = fl_18_df["County"].str.contains("Number")

filter_num_true = filter_number.loc[filter_number].index

filter_num_plus = filter_number.loc[filter_number].index+1

value = fl_18_df.loc[filter_num_plus,["2018 Median Income"]]

fl_18_df.head(15)
```

```python
#Moving Median Income Values
fl_18_df['number'] = fl_18_df["2018 Median Income"].where(fl_18_df.index.isin(filter_num_plus))

fl_18_df['number'] = fl_18_df['number'].ffill()

#fl_18_df = fl_18_df.loc[~fl_18_df['2018 Median Income']== np.nan]

fl_18_df = fl_18_df.loc[~fl_18_df['2018 Median Income'].isnull()]

fl_18_df.head(15)
```

```python
#Drop unnecessary columns
fl_18_df = fl_18_df.reset_index()

fl_18_df = fl_18_df.drop(["index","County","number"], axis=1)

fl_18_df.head(15)
```

```python
#Only keep every third row (actual median income value)

fl_18_df = fl_18_df.iloc[2::3]

fl_18_df.head(15)
```

```python
#Only keep county names
fl_18_df["county"] = fl_18_df["county"].str.replace(", Florida", "")

fl_18_df.head(10)
```

```python
#Reset index and rename column
fl_18_df = fl_18_df.reset_index(drop=True)

fl_18_df.rename(columns={"county":"County"},inplace=True)

fl_18_df.head(15)
```

```python
fl_Clean18_df = fl_18_df.set_index("County")

fl_Clean18_df.head(10)
```
