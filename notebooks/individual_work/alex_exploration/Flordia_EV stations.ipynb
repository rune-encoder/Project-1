```python
zip_df=pd.read_csv(
    Path("../../../data/processed_data/fl_zipcodes.csv"),
    dtype={"Zip Code":str}
)
zip_df
```

```python
zip_df.info()
```

```python
zip_group_df=zip_df.groupby("County")["Zip Code"].apply(list)
zip_group_df
```

```python
zip_map=zip_group_df.to_dict()
#zip_map
```

```python
def map_zip_to_county(zip_code):
    for key, value in zip_map.items():
        if zip_code in value: 
            return key, value 
    return None, None

            
```

```python
ev_stations_clean['ev_station_count'] = 1  # Create a new column with a default value of 1
ev_stations_per_year = ev_stations_clean.groupby(["year", "county"])["ev_station_count"].sum().reset_index()
```

```python
ev_stations_per_year = ev_stations_clean.groupby(["year","county"])["ev_station_count"].sum().reset_index()
ev_stations_per_year
```

```python
ev_stations_per_year.sort_values(by=["county","year"])
```

```python
ev_stations_per_year["cumulative_ev_stations"]= ev_stations_per_year.groupby("county")["ev_station_count"].cumsum()
ev_stations_per_year.sort_values(by=["county","year"])
```

```python
# Create a function to save the DataFrames to CSV files
def save_csv_file(df, file_path):
    """ 
    Save a DataFrame to a CSV file at the specified file path.

    Parameters:
    - df: DataFrame to save
    - file_path: Path to save the CSV file
    """
    
    # Check if the parent directory exists
    if not file_path.parent.exists():
        print(f"Error: The directory `{file_path.parent}` does not exist.")
        return
    
    if file_path.exists():
        print(f"File `{file_path.name}` already exists. Overwriting file.")
        file_path.unlink()
    
    # Save the DataFrame to the specified file path
    df.to_csv(file_path, index=False)
    print(f"File saved as `{file_path.name}`")
```

```python
file_name = "fl_evstation_progression.csv"
file_path = Path(f"../../../data/processed_data/{file_name}")
save_csv_file(ev_stations_per_year, file_path)
```
